```python
import csv
tweets = []
with open('files/clean_enero.csv', encoding="utf8") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        line_count += 1
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            tweets.append(row[0])
            line_count += 1
    print(f'Processed {line_count} lines.')
```

```python
from generateWordCloud import generateWordCloud
```

```python
cloud = generateWordCloud(tweets)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(cloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show() 
```

En lo que respecta a la nube de palabraas de enero podemos observar que los temas se centran alrededor del nuevo gobierno de Alejandro Giammattei. Esto se confirma con palabras como, presidente electo, nombramiento, casa presidencial, posesión, etc.
